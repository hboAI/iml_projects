```python
from pathlib import Path
import tensorflow as tf
from tensorflow.keras.callbacks import TensorBoard
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import torch
from torch.utils.data import DataLoader, TensorDataset
import torch.nn as nn
import torch.nn.functional as F
```

```python
train_data = np.load("train_data.npz")["data"]
image_1 = train_data[0].squeeze()
mask = image_1.copy()
mask[10:18, 10:18] = 0
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image_1, cmap="gray")
axes[0].set_title("Original Image")
axes[0].axis('off')


axes[1].imshow(mask, cmap="gray")
axes[1].set_title("Masked Image")
axes[1].axis('off')

plt.show()
```

```python
def apply_mask(image):
    """
    Apply a mask to an image.
    """
    masked_image = image.copy()
    masked_image[10:18, 10:18] = 0  # Example masking operation
    return masked_image


masked_data = np.array([apply_mask(image.squeeze()) for image in train_data])


plt.imshow(masked_data[0].squeeze(), cmap="gray")
plt.title("Masked Image")
plt.axis('off')  # Turn off axis for a cleaner look
plt.show()
```
